# Ejecución de celdas de código

En este apartado se muestra cómo declarar y ejecutar celdas de código en un Jupyter Book y poner ciertas variables a disposición de otros Jupyter Notebook del libro.

**Requisitos**:
- El usuario debe estár abierto a modificar celdas de código en un Jupyter Notebook.
- El Jupyter Book debe construirse en HTML según lo indicado en el {doc}`Apéndice <apendice1>`

A continuación se muestran dos variables de entrada _a, b_ y una de salida _c_:

````python
a = 1
b = 2
c = a + b
````

El resultado puede indicarse en una celda de código con etiqueta _hide-input_ para evitar mostrar el código que lo genera:


````python
print('Resultado:', c)
````

En otra celda de código con etiqueta _hide-input_, se escribe el código que guarda las variables relevantes (en este caso _c_):
```python
import contextlib
import io
with contextlib.redirect_stdout(io.StringIO()):
    %store c
```  
Esta variables estará disponible para siguientes Jupyter Notebook (véase el Notebook que sigue como ejemplo).

````python
import contextlib
import io
with contextlib.redirect_stdout(io.StringIO()):
    %store c
````
