# Procesar algoritmos en LaTeX

En este apartado se presentan las bases para la creación del paquete **compylatex**. La idea inicial es escribir un algoritmo completo en LaTeX. En este caso, en una celda Markdown como la siguiente.

$$
\begin{array}{l}
\textbf{Entrada: } x_0, f, f' \\
\textbf{Salida: } x^* \text{ tal que } f(x^*) \approx 0 \\
\hline
\text{Inicializar } x \gets x_0 \\
\text{Para } k = 0,1,2,\dots \text{ hacer:} \\
\quad x_{\text{nuevo}} \gets x - \frac{f(x)}{f'(x)} \\
\quad \text{Si } |x_{\text{nuevo}} - x| < \varepsilon \text{ entonces: salir} \\
\quad x \gets x_{\text{nuevo}} \\
\text{Fin Para} \\
x^* \gets x
\end{array}
$$

El siguiente paso, consistiría en leer dicho algoritmo en una celda de código, como ya hemos visto en un {doc}`apartado anterior <acceso1>`.

Debemos definir con cuidado qué entornos de tipo algoritmo serán válidos en nustro código [ver página de explicación de los distintos paquetes LaTeX para algoritmos](https://tex.stackexchange.com/questions/229355/algorithm-algorithmic-algorithmicx-algorithm2e-algpseudocode-confused).

Además, debemos identificar un paquete Python que permita procesar dicho algoritmo. En este trabajo se ha optado por [pylatexenc](https://pypi.org/project/pylatexenc/). Sin embargo, este paquete analiza documentos LaTeX completos (no fórmulas Latex en celdas de Jupyter Notebook), por lo que en el apartado corrspondiente al paquete **compylatex** desechamos el acceso a celdas, descrito en estos Preliminares.
