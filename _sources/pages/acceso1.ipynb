# Acceso a otras celdas del Notebook

En este apartado mostramos cómo acceder desde una celda de código al contenido de otra celda Markdown. Este ejemplo ilustraría el caso en que un usuario quisiese escribir sus fórmulas LaTeX en una celda Markdown para que el código de cálculo, que el usuario no necesita tocar, ejecutase dicha fórmula.

En este primer caso, se supone que se conoce el índice absoluto de la celda de código que se está ejecutando. Esta situación no es recomendable si el IPYNB está cambiando, por ser demasiado rígida. A continuación, las dos celdas: Mardown y de código. Se supone además que la primera sigue a la segunda.

$\frac{1 + \sqrt{a}}{b}+2$

```python
#import ipynbname
import nbformat

# Obtener el notebook actual
#notebook_path = ipynbname.path()

with open('./acceso1.ipynb', 'r', encoding='utf-8') as f:
    nb = nbformat.read(f, as_version=4)

# Suponiendo que esta es la celda 3 (índice 2)
current_cell_index = 3  # Se conoce qué celda se está ejecutando
previous_cell_index = current_cell_index - 1

if previous_cell_index >= 0:
    previous_cell = nb.cells[previous_cell_index]
    print(f"Tipo: {previous_cell.cell_type}")
    print(f"Contenido: {previous_cell.source}")
else:
    print("No hay celda anterior")
```

En este segundo ejemplo, más flexible, la celda Markdown se lee en la celda de código porque la primera tiene la etiqueta **"formula"**. Esta idea se adaptará para las etiquetas de LaTeX en el paquete **compylatex**.

$\frac{1 + \sqrt{a}}{b}+3$

```python
#import ipynbname
import nbformat
import json

def leer_celdas_con_tag(tag_buscado):
    """Encuentra celdas markdown con un tag específico"""
    
    with open('./acceso1.ipynb', 'r', encoding='utf-8') as f:
        nb = nbformat.read(f, as_version=4)
    
    celdas_encontradas = []
    for i, cell in enumerate(nb.cells):
        # Verificar si tiene metadatos y tags
        if (cell.cell_type == 'markdown' and 
            'metadata' in cell and 
            'tags' in cell.metadata):
            
            if tag_buscado in cell.metadata['tags']:
                celdas_encontradas.append({
                    'index': i,
                    'content': cell.source
                })
    
    return celdas_encontradas

# Uso
celdas_markdown = leer_celdas_con_tag("formula")
for celda in celdas_markdown:
    print(f"Celda {celda['index']}: {celda['content']}")
```
